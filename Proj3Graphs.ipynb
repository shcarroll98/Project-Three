```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('C:/Users/secar/OneDrive/Documents/Bootcamp/sp500_merged.csv')
```

```python
#Filter out stock prices over 1000
df_filtered = df[df['Adj Close'] <= 1000]

# Extract 'Adj Close' and 'total_score' columns
adj_close = df_filtered['Adj Close']
total_score = df_filtered['total_score']

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=total_score, y=adj_close)
plt.title('ESG Score vs. Stock Prices')
plt.xlabel('Total ESG Score')
plt.ylabel('Adj Close Stock Price')
plt.grid(True)
plt.show()
```

```python
# Plotting with reversed axes
plt.figure(figsize=(10, 6))

sns.scatterplot(x=adj_close, y=total_score)
plt.title('Stock Prices vs. ESG Score')
plt.xlabel('Adj Close Stock Price')
plt.ylabel('Total ESG Score')
plt.grid(True)
plt.show()
```

```python
# Filter out volumes greater than 20,000,000
volume_filtered = df[df['Volume'] <= 20000000]

# Extract 'Volume' and 'total_score' columns
volume = volume_filtered['Volume']
total_score = volume_filtered['total_score']


# Plotting with reversed axes
plt.figure(figsize=(10, 6))
#plt.plot(volume, total_score, marker='o', linestyle='-')
sns.scatterplot(x=volume, y=total_score)
plt.title('Volume vs. ESG Score')
plt.xlabel('Volume')
plt.ylabel('Total ESG Score')
plt.grid(True)
plt.show()
```
